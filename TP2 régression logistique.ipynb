```python
def sigmoid(z):
    return ( 1/(1+np.exp(-z)) )
```

```python
def cost_function(x , y , theta):
    m = y.size
    eps = 1e-9
    h = sigmoid(x.dot(theta))
    h = np.clip(h, eps, 1-eps)
    J = (1/m)*np.sum(-y*np.log(h) - (1-y)*np.log(1-h))
    return J
```

```python
def gradient_descent(x, y, theta, alpha):
    m = y.size
    J = np.zeros(200100)
    for i in range(200100):
        h = sigmoid(x.dot(theta))
        theta = theta - alpha*(1/m)*((h - y).dot(x))
        J[i] = cost_function(x, y, theta)
    return theta, J
```

```python
def normalize(x):
    x_norm = (x - np.min(x, axis=0)) / (np.max(x, axis=0) - np.min(x, axis=0))
    return x_norm
```

```python
x = np.c_[np.ones(x.shape[0]), x]
```

```python
# définir les hyperparamètres du modèle
alpha = 0.01
theta = np.zeros(x.shape[1])
# exécuter la descente de gradient pour estimer les paramètres du modèle
theta, J = gradient_descent(x, y,theta,alpha)
print('Paramètres du modèle :',theta[0],theta[1],theta[2] , "J =", J[-1])
```
