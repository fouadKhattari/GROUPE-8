```python
sns.pairplot(df, vars = ['Spending Score (1-100)', 'Annual Income (k$)', 'Age'], hue = "Genre")
```

```python
datasubset = df.loc[:, ["Spending Score (1-100)", "Annual Income (k$)"]]
```

```python
plt.figure(figsize=(9, 6))
plt.scatter(datasubset[["Annual Income (k$)"]], datasubset[["Spending Score (1-100)"]],c='blue')
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10, 7))


dendrogram = sch.dendrogram(sch.linkage(datasubset, method  = "ward"))
plt.title('Dendrogramme')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()

#vrifier la plus grande distance verticalement sans traverser aucune ligne horizontale
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')  
cluster.fit_predict(datasubset)
```

```python
cl = cluster.fit_predict(datasubset)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(datasubset,cl)
```

```python
X = datasubset.values
X
```

```python
plt.figure(figsize=(9, 6))  
plt.scatter(X[cl==0, 0], X[cl==0, 1], s=100, c='black', label ='Cluster 1')
plt.scatter(X[cl==1, 0], X[cl==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(X[cl==2, 0], X[cl==2, 1], s=100, c='green', label ='Cluster 3')
plt.scatter(X[cl==3, 0], X[cl==3, 1], s=100, c='yellow', label ='Cluster 4')
plt.scatter(X[cl==4, 0], X[cl==4, 1], s=100, c='red', label ='Cluster 5')
plt.title('Clusters of Mall Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
