```python
# ajouter une colonne de 1 pour le terme d'interception
X = np.c_[np.ones(X.shape[0]), X]
```

```python
X
```

```python
# initialiser les paramètres du modèle à zéro
theta = np.zeros(2)
theta
```

```python
# définir la fonction de coût
def compute_cost(X, y, theta):
    m = y.size
    h = X.dot(theta)
    J = 1/(2*m)*np.sum((h-y)**2)
    return J
```

```python
# définir la fonction de gradient de la fonction de coût
def gradient_descent(X, y, theta, alpha, num_iters):
    m = y.size
    J_history = np.zeros(num_iters)
    for i in range(num_iters):
        h = X.dot(theta)
        theta = theta - alpha*(1/m)*(X.T.dot(h-y))
        J_history[i] = compute_cost(X, y, theta)
    return theta, J_history
```

```python
# définir les hyperparamètres du modèle
alpha = 0.01
num_iters = 1500
```

```python
# exécuter la descente de gradient pour estimer les paramètres du modèle
theta, J_history = gradient_descent(X, y, theta, alpha, num_iters)
print('Paramètres du modèle :', theta)
```

```python
# tracer la courbe d'apprentissage
plt.plot(np.arange(num_iters), J_history)
plt.xlabel('Nombre d\'itérations')
plt.ylabel('Fonction de coût')
plt.show()
```

```python
# tracer la droite de régression linéaire sur les données
plt.scatter(X[:, 1], y, marker='x', color='r')
plt.plot(X[:, 1], X.dot(theta))
plt.xlabel('Population de la ville en 10 000')
plt.ylabel('Bénéfice en $10 000')
plt.show()
```
